## Create Files for Use in Text Classification and Sentiment Analysis
The definition below uses the spaCy and nltk libraries to create two pickle files for use in the Text and Sentiment Analysis portion of the project. We are creating separate pickle files for this portion of the project because in earlier tests, we found that initial pickle files were accurate for prediction, but removed too much information for accurate topic modeling.

```python
# function to preprocess reviews.
def preprocess_reviews(df_raw, stop_words: set, eng_words: set):
    """Perform lemmatization, remove stopwords, punctuation, non-english words."""
    def delete_words(row): 
        """Delete stop words, punctuation, and non-english words from a review."""
        row = row.split()
        result_row = [word for word in row 
                      if word not in stop_words
                      and word in eng_words]
        return result_row
    
    df = df_raw.copy()
    
    # Lowercase reviews
    df['review'] = df.apply(lambda row: str(row['review']).lower(), axis=1)
    
    # Lemmatization
    df['review'] = df['review'].apply(lambda row: " ".join([w.lemma_ for w in nlp(row)]))
    
    # Delete stop words and non-english words
    df['review'] = df['review'].apply(delete_words)
    
    # Join list of tokens
    df['review'] = df['review'].apply(lambda row: " ".join(row))
    
    return df
```

```python

# Load stop words
stop_words = nlp.Defaults.stop_words

# Add punctuation to stop words
punctuation = string.punctuation
stop_words.update(set(punctuation))

# Add name of device to stop words
stop_words.add('echo')
stop_words.add('dot')


# Load english words
eng_words = set(nltk.corpus.words.words())
```

```python
%%time
w_corpus17_clean = preprocess_reviews(w_corpus17, stop_words, eng_words)
w_corpus18_clean = preprocess_reviews(w_corpus18, stop_words, eng_words)
```

```python
# Save preprocessed data to pickle files
w_corpus17_clean.to_pickle("../data/df_text_train.pickle")
w_corpus18_clean.to_pickle("../data/df_text_test.pickle")
```

## Build a pre-model
By using a pre-model, we can get a rough estimatation of how our model will perform and fine tune it before using it on the test data.

```python
# Create two objects X and y. X will be the 'review' column of the training set
# and y will be the 'rating' column.
X = df_train['review']
y = df_train['rating']
```

With CountVectorizer, we are using a bag-of-words approach to the representation of the corpora. It disregards the structure of the text and simply counts how often each word appears in the corpus.

```python
# make a count vectorizer object to tokenize the training data and build a vocabulary
cv = CountVectorizer()
```

```python
# use fit_transform on CountVectorizer and pass in X
X = cv.fit_transform(X)
```

```python
# Let's see the vocabulary size with the countvectorizer
print("Vocabulary size: {}".format(len(cv.vocabulary_)))
```

```python
# Let's see the vocabulary content with the countvectorizer
print("Vocabulary content:\n {}".format(cv.vocabulary_))
```

```python
# visualize the dictionary
cvdictionary = cv.vocabulary_
counts = list(cvdictionary.values())
words = list(cvdictionary.keys())
cvdictdf = pd.DataFrame(words, columns = ['words'])
cvdictdf['counts'] = counts
cvdictdf.sort_values(by=['counts'], inplace=True, ascending=False)
print(cvdictdf)
```

## Train Test Split for the Pre-Model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                test_size=0.3, random_state=101)
```

## Train the Pre-Model

Naive Bayes classifiers are efficient in supervised learning because they look at each feature and collect simple per-class statistics from each feature. 
As a Naive Bayes classifier, MultinomialNB is useful for text data classification because it examines count data (each feature (word) represents how often it appears in a corpora), specifically the average value for each feature in each class. 

```python
# create an instance of the Naive Bayes estimator
nb = MultinomialNB()
```

```python
# fit nb using the training data
nb.fit(X_train, y_train)
```

## Predictions and Evaluations of the Pre-Model

```python
# use the predict method off of nb to predict labels from X_test
predictions = nb.predict(X_test)
```

```python
# create a confusion matrix and classification report 
# using these predictions and y_test
print(confusion_matrix(y_test, predictions))
print('\n')
print(classification_report(y_test,predictions))
```

Because we included all ratings our accuracy is only 68%. However, our F1-Scores were highest for the 1 and 5 star ratings, indicating a potentially optimal blend of precision and recall with these ratings.\s\s

**Let's think how the model could be improved:**
- Classifying the reviews by rating. If we think about the ratings, we know 1 is a poor rating, 3 is an average rating, and 5 is an excellent rating. If we filter the training corpus to only use 1 and 5 star ratings, we are creating a more polar basis for assessment.
- Count Vectorization (CountVectorizer()) counts the occurances (token counts) of all the unique words in each document. However, TfIDF (TfidfVectorizer()) calculates the term frequency-inverse document frequency value for each word as it relates to the document term matrix. 
- MultinomialNB() can be used to predict the tag of a text based on the probability of each tag for a given text and then outputing the tag with the highest one.

## Topic Modeling
A form of unsupervised learning that clusters documents together by topic.

### Latent Dirichlet Allocation (LDA)
Uses the Dirichlet Probability Distribution to group documents with similar words. Latent topics can be found by searching for words that occur toegether. The topics are probablility distributions over the words.

We're going to base our topic model off of the ratings and our positivity discovery found during our EDA. Therefore, we will create two sets of topic models: one where k=2, one where k=3, and one where k=5.
